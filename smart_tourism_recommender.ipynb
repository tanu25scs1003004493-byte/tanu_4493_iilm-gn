## Cell 1 – Library Imports

```python
# Cell 1 – Library Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

print("✅ All libraries imported successfully!")
```

## Cell 2 – User Preferences & Real-Time Tourism Data

```python
# Cell 2 – User Preferences & Real-Time Tourism Data

# Simulated user input
user_preferences = {
    "budget": "medium",
    "interests": ["nature", "heritage"],
    "travel_month": "December",
    "location": "India"
}

# Simulated real-time destination data
realtime_data = {
    "destinations": [
        {"name": "Goa",     "crowd_level": "high",   "weather": "sunny", "type": "beach"},
        {"name": "Jaipur",  "crowd_level": "medium", "weather": "clear", "type": "heritage"},
        {"name": "Kerala",  "crowd_level": "low",    "weather": "rainy", "type": "nature"},
        {"name": "Manali",  "crowd_level": "medium", "weather": "snowy", "type": "nature"},
        {"name": "Varanasi","crowd_level": "high",   "weather": "clear", "type": "heritage"},
        {"name": "Coorg",   "crowd_level": "low",    "weather": "cloudy","type": "nature"}
    ]
}

# Historical monthly visitor count (last 6 months)
historical_visitors = [1200, 1500, 1700, 2000, 1800, 2100]

print("✅ Data loaded!")
print(f"User Interests : {user_preferences['interests']}")
print(f"Travel Month   : {user_preferences['travel_month']}")
print(f"Budget         : {user_preferences['budget']}")
print(f"\nAvailable Destinations:")
for d in realtime_data["destinations"]:
    print(f"  - {d['name']:10s} | Type: {d['type']:8s} | Crowd: {d['crowd_level']:6s} | Weather: {d['weather']}")
```

## Cell 3 – Recommendation Module

```python
# Cell 3 – Recommendation Module
# Filters destinations matching user interests and avoids high-crowd places

def generate_recommendations(user_prefs, data):
    """
    Recommends destinations based on:
    - Matching at least one user interest to destination type
    - Excluding high-crowd destinations
    """
    recommendations = []
    user_interests = user_prefs["interests"]

    for dest in data["destinations"]:
        # Check if destination type matches any user interest
        interest_match = dest["type"] in user_interests
        # Filter out high-crowd destinations
        crowd_ok = dest["crowd_level"] != "high"

        if interest_match and crowd_ok:
            recommendations.append({
                "name":        dest["name"],
                "type":        dest["type"],
                "crowd_level": dest["crowd_level"],
                "weather":     dest["weather"]
            })

    return recommendations


# Run recommendations
recommendations = generate_recommendations(user_preferences, realtime_data)

print("=== 🏖️ RECOMMENDED DESTINATIONS ===")
if recommendations:
    for i, r in enumerate(recommendations, 1):
        print(f"{i}. {r['name']:10s} | Type: {r['type']:8s} | Crowd: {r['crowd_level']:6s} | Weather: {r['weather']}")
else:
    print("No matching destinations found. Try broadening your interests.")
```

## Cell 4 – Demand Prediction Module (Linear Regression)

```python
# Cell 4 – Demand Prediction Module using Linear Regression

def predict_demand(history):
    """
    Predicts next month's visitor count using Linear Regression.
    Input  : list of historical monthly visitor counts
    Output : integer prediction for the next month
    """
    model = LinearRegression()
    X = np.arange(len(history)).reshape(-1, 1)   # Month indices [0,1,2,...]
    y = np.array(history)                         # Visitor counts

    model.fit(X, y)

    # Predict the next month
    next_month = np.array([[len(history)]])
    prediction = int(model.predict(next_month)[0])

    # Training accuracy (MAE)
    y_pred_train = model.predict(X)
    mae = mean_absolute_error(y, y_pred_train)

    return prediction, mae, model


# Run demand prediction
demand_forecast, mae, lr_model = predict_demand(historical_visitors)

print("=== 📊 DEMAND PREDICTION MODULE ===")
print(f"Historical Data  : {historical_visitors}")
print(f"Model            : Linear Regression")
print(f"Training MAE     : {mae:.2f} visitors")
print(f"\n📅 Predicted Visitor Flow (Next Month): {demand_forecast} visitors")

# Identify peak / off-peak
avg = np.mean(historical_visitors)
if demand_forecast > avg:
    print("📈 Status: PEAK PERIOD – Prepare for high footfall")
else:
    print("📉 Status: OFF-PEAK PERIOD – Good for budget travelers")
```

## Cell 5 – Combined System Output

```python
# Cell 5 – Combined System Output (Dashboard Print)

output = {
    "user_preferences": user_preferences,
    "recommended_destinations": [r["name"] for r in recommendations],
    "predicted_visitor_flow_next_month": demand_forecast,
    "demand_status": "PEAK" if demand_forecast > np.mean(historical_visitors) else "OFF-PEAK"
}

print("\n" + "="*45)
print("      🌐 SMART TOURISM SYSTEM OUTPUT")
print("="*45)
print(f"  Budget            : {output['user_preferences']['budget']}")
print(f"  Interests         : {', '.join(output['user_preferences']['interests'])}")
print(f"  Travel Month      : {output['user_preferences']['travel_month']}")
print("-"*45)
print(f"  Recommended Spots : {', '.join(output['recommended_destinations'])}")
print(f"  Next Month Flow   : {output['predicted_visitor_flow_next_month']} visitors")
print(f"  Demand Status     : {output['demand_status']}")
print("="*45)
```

## Cell 6 – Visualization

```python
# Cell 6 – Visualization

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Smart Tourism System – Results Dashboard", fontsize=14, fontweight='bold')

# ── Plot 1: Visitor Demand Forecast ──────────────────────────────────
ax1 = axes[0]
months = [f"Month {i+1}" for i in range(len(historical_visitors))]
months_all = months + ["Next Month"]
visitors_all = historical_visitors + [demand_forecast]

colors = ['#4C72B0'] * len(historical_visitors) + ['#DD8452']
bars = ax1.bar(months_all, visitors_all, color=colors, edgecolor='white', linewidth=0.8)

# Trend line over historical data
x_vals = np.arange(len(historical_visitors)).reshape(-1, 1)
trend = lr_model.predict(np.arange(len(visitors_all)).reshape(-1, 1))
ax1.plot(months_all, trend, 'r--', linewidth=1.5, label='Trend Line')

for bar, val in zip(bars, visitors_all):
    ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 30,
             str(val), ha='center', va='bottom', fontsize=9)

hist_patch  = mpatches.Patch(color='#4C72B0', label='Historical')
pred_patch  = mpatches.Patch(color='#DD8452', label='Predicted')
ax1.legend(handles=[hist_patch, pred_patch], fontsize=9)
ax1.set_title("Monthly Visitor Flow & Forecast")
ax1.set_ylabel("Visitor Count")
ax1.set_ylim(0, max(visitors_all) * 1.2)
ax1.tick_params(axis='x', rotation=30)
ax1.grid(axis='y', linestyle='--', alpha=0.5)

# ── Plot 2: Recommended Destinations (Crowd Level) ───────────────────
ax2 = axes[1]
crowd_map   = {"low": 1, "medium": 2, "high": 3}
color_map   = {"low": "#2ecc71", "medium": "#f39c12", "high": "#e74c3c"}

all_dests   = realtime_data["destinations"]
dest_names  = [d["name"] for d in all_dests]
crowd_vals  = [crowd_map[d["crowd_level"]] for d in all_dests]
bar_colors  = [color_map[d["crowd_level"]] for d in all_dests]
rec_names   = [r["name"] for r in recommendations]

bars2 = ax2.barh(dest_names, crowd_vals, color=bar_colors, edgecolor='white')

for bar, name in zip(bars2, dest_names):
    marker = "  ✅ Recommended" if name in rec_names else ""
    ax2.text(bar.get_width() + 0.05, bar.get_y() + bar.get_height()/2,
             marker, va='center', fontsize=9, color='green')

ax2.set_xticks([1, 2, 3])
ax2.set_xticklabels(["Low", "Medium", "High"])
ax2.set_xlabel("Crowd Level")
ax2.set_title("Destination Crowd Levels")
ax2.set_xlim(0, 4.5)

low_p  = mpatches.Patch(color='#2ecc71', label='Low')
med_p  = mpatches.Patch(color='#f39c12', label='Medium')
high_p = mpatches.Patch(color='#e74c3c', label='High')
ax2.legend(handles=[low_p, med_p, high_p], title="Crowd", fontsize=8)

plt.tight_layout()
plt.savefig("smart_tourism_results.png", dpi=150, bbox_inches='tight')
plt.show()
print("✅ Chart saved as smart_tourism_results.png")
```
